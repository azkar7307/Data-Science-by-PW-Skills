# Q1. 

What is Matplotlib? Why is it used? Name five plots that can be plotted using the Pyplot module of Matplotlib.


## Answer


Matplotlib is a data visualization library for Python that allows you to create a wide range of static, animated, and interactive visualizations in Python. It is widely used in scientific computing, data science, and machine learning for creating high-quality plots, histograms, bar charts, scatterplots, and more.

Matplotlib's Pyplot module provides a simple interface for creating and customizing various types of plots. Here are five examples of plots that can be created using the Pyplot module:

* `Line plot`: A line plot is used to display a continuous relationship between two variables. It can be created using the plot() function.

* `Bar chart`: A bar chart is used to compare different categories of data. It can be created using the bar() function.

* `Scatter plot`: A scatter plot is used to display the relationship between two variables. It can be created using the scatter() function.

* `Histogram`: A histogram is used to display the distribution of a single variable. It can be created using the hist() function.

* `Pie chart`: A pie chart is used to show the proportion of each category in a dataset. It can be created using the pie() function.




# Q2. 

What is a scatter plot? Use the following code to generate data for x and y. Using this generated data plot a scatter plot.


```python
np.random.seed(3)
x = 3 + np.random.normal(0, 2, 50)
y = 3 + np.random.normal(0, 2, len(x))
```

Note: Also add title, xlabel, and ylabel to the plot.


## Answer

A scatter plot is a type of plot that displays the relationship between two continuous variables. It is used to show how one variable is affected by another variable. Each point in the plot represents a single observation, with the x-coordinate representing the value of one variable, and the y-coordinate representing the value of the other variable.

Here's the code to generate the data and create a scatter plot using Matplotlib's Pyplot module:

```python
print(x)
print("-"*30)
print(y)
```

```python
plt.scatter(x, y)
plt.title("Scatter Plot of x and y")
plt.xlabel("X values")
plt.ylabel("Y values")
plt.show()
```

In this code, we first import the required libraries, generate the data for x and y using NumPy's random module, and then create the scatter plot using the `scatter()` function of the Pyplot module. We also add a title to the plot using the `title()` function, and label the x and y axes using the `xlabel()` and `ylabel()` functions, respectively.

When we run this code, it will generate a scatter plot of x and y, with x values on the x-axis and y values on the y-axis. The plot will also have a title and labeled axes.

Why is the subplot() function used? Draw four line plots using the subplot() function.

# *Use the following data:*

import numpy as np

For line 1: x = np.array([0, 1, 2, 3, 4, 5]) and y = np.array([0, 100, 200, 300, 400, 500])

For line 2: x = np.array([0, 1, 2, 3, 4, 5]) and y = np.array([50, 20, 40, 20, 60, 70])

For line 3: x = np.array([0, 1, 2, 3, 4, 5]) and y = np.array([10, 20, 30, 40, 50, 60])

For line 4: x = np.array([0, 1, 2, 3, 4, 5]) and y = np.array([200, 350, 250, 550, 450, 150])


## Answer

The `subplot()` function in Matplotlib is used to create multiple plots in the same figure. It allows us to display multiple plots in a grid format, where each plot is a separate axis within the figure. This function takes three arguments: the number of rows, the number of columns, and the index of the subplot we want to create.

Here's the code to create four line plots using the `subplot()` function in Matplotlib:

```python
# Data for line 1
x1 = np.array([0, 1, 2, 3, 4, 5])
y1 = np.array([0, 100, 200, 300, 400, 500])

# Data for line 2
x2 = np.array([0, 1, 2, 3, 4, 5])
y2 = np.array([50, 20, 40, 20, 60, 70])

# Data for line 3
x3 = np.array([0, 1, 2, 3, 4, 5])
y3 = np.array([10, 20, 30, 40, 50, 60])

# Data for line 4
x4 = np.array([0, 1, 2, 3, 4, 5])
y4 = np.array([200, 350, 250, 550, 450, 150])
```

```python
# Create a 2x2 grid of subplots
fig, axs = plt.subplots(2, 2)

# Plot data on each subplot
axs[0, 0].plot(x1, y1)
axs[0, 1].plot(x2, y2)
axs[1, 0].plot(x3, y3)
axs[1, 1].plot(x4, y4)

# Add titles to each subplot
axs[0, 0].set_title("Line 1")
axs[0, 1].set_title("Line 2")
axs[1, 0].set_title("Line 3")
axs[1, 1].set_title("Line 4")

# Add overall title to the figure
fig.suptitle("Four Line Plots")

# Display the plot
plt.show()
```

In this code, we first import the required libraries and define the data for each line plot using NumPy's `array()` function. We then create a 2x2 grid of subplots using the `subplots()` function, and assign the resulting axes to the variable `axs`.

We then plot each line on its corresponding subplot using the `plot()` function of the appropriate axis object. We add titles to each subplot using the `set_title()` function, and add an overall title to the figure using the `suptitle()` function.

When we run this code, it will generate a figure with four line plots, arranged in a 2x2 grid. Each plot will have its own title, and there will be an overall title for the entire figure.

# Q4. 

What is a bar plot? Why is it used? Using the following data plot a bar plot and a horizontal bar plot.


```python
company = np.array(["Apple", "Microsoft", "Google", "AMD"])
profit = np.array([3000, 8000, 1000, 10000])
```

## Answer


A bar plot is a graphical representation of categorical data with rectangular bars, where the height or length of each bar corresponds to the value of a particular category. Bar plots are used to compare the magnitudes of different categories or groups of data.

A horizontal bar plot, also known as a bar chart, is a bar plot where the bars are horizontal instead of vertical.

Here's how to plot a bar plot and a horizontal bar plot for the given data in Matplotlib:

```python
# Create a bar plot
plt.bar(company, profit)

# Add title and axis labels
plt.title("Company Profit")
plt.xlabel("Company")
plt.ylabel("Profit")

# Display the plot
plt.show()
```

This code will produce a bar plot with the x-axis representing the company names and the y-axis representing the profits of each company.

To create a horizontal bar plot, we simply need to use the `barh()` function instead of the `bar()` function:

```python
# Create a horizontal bar plot
plt.barh(company, profit)

# Add title and axis labels
plt.title("Company Profit")
plt.xlabel("Profit")
plt.ylabel("Company")

# Display the plot
plt.show()
```

This code will produce a horizontal bar plot with the y-axis representing the company names and the x-axis representing the profits of each company.

# Q5. 

What is a box plot? Why is it used? Using the following data plot a box plot.


```python
box1 = np.random.normal(100, 10, 200)
box2 = np.random.normal(90, 20, 200)
```

## Answer

A box plot, also known as a box-and-whisker plot, is a graphical representation of the distribution of data based on the five-number summary: the minimum value, the first quartile (Q1), the median (Q2), the third quartile (Q3), and the maximum value. Box plots are used to identify the median, the range, the variability and outliers in the data.

Here's how to plot a box plot for the given data in Matplotlib:

```python
# Combine the two data sets
data = [box1, box2]
data
```

```python
print(type(data))
print(len(data))
```

```python
print(len(data[0]))
print(len(data[1]))
```

```python

# Create a box plot
plt.boxplot(data)

# Add title and axis labels
plt.title("Box Plot")
plt.xlabel("Box")
plt.ylabel("Value")

# Display the plot
plt.show()
```

This code will produce a box plot with two boxes, one for `box1` and another for `box2`. The x-axis represents the boxes, and the y-axis represents the values. The box itself represents the interquartile range (IQR), the line inside the box represents the median, and the whiskers represent the minimum and maximum values of the data that are not outliers. Outliers are shown as individual points outside the whiskers.
