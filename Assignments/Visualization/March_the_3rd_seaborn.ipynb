# Q1. 

Name any five plots that we can plot using the Seaborn library. Also, state the uses of each plot.


## Answer

Here are five plots that can be plotted using Seaborn library along with their uses:

* `Scatter plot` - Scatter plot is used to plot the distribution of two continuous variables. It is useful to understand the relationship between two variables and identify any trends, clusters, or outliers.

* `Line plot` - A line plot is used to plot a continuous variable over time or a sequence. It is useful to visualize trends and changes over time.

* `Bar plot` - A bar plot is used to plot categorical data with rectangular bars. It is useful to compare different categories and their values.

* `Histogram` - A histogram is used to visualize the distribution of a continuous variable. It is useful to identify the central tendency, spread, and skewness of the data.

* `Heatmap` - A heatmap is used to plot a matrix of values with colors that represent their magnitude. It is useful to visualize correlation matrices or confusion matrices.

Seaborn library provides various customization options to these plots like adding titles, labels, legends, colors, and themes to make them visually appealing and informative.

# Q2.

Load the "fmri" dataset using the load_dataset function of seaborn. Plot a line plot using x = "timepoint" and y = "signal" for different events and regions.

Note: timepoint, signal, event, and region are columns in the fmri dataset.


## Answer

Here's the code to load the "fmri" dataset and plot a line plot using x = "timepoint" and y = "signal" for different events and regions:

```python
# Load the fmri dataset
fmri = sns.load_dataset("fmri")
fmri.head()
```

```python
fmri.info()
```

```python
# Plot a line plot for different events and regions
sns.lineplot(x="timepoint", y="signal", hue="event", style="region", data=fmri)
```

This code uses the `load_dataset()` function from Seaborn to load the "fmri" dataset. Then, the `lineplot()` function is used to create a line plot with x-axis as "timepoint" and y-axis as "signal". The "event" column is used to group the data points by different events, and the "region" column is used to plot different line styles for different regions.

The resulting plot will show the trend of the signal over time for each event and region combination. The hue parameter sets the colors of the plot, while the style parameter sets the different line styles for each region.

# Q3. 

Load the "titanic" dataset using the load_dataset function of seaborn. Plot two box plots using x = 'pclass', y = 'age' and y = 'fare'.

Note: pclass, age, and fare are columns in the titanic dataset.


## Answer

Here's the code to load the "titanic" dataset and plot two box plots using x = 'pclass', y = 'age' and y = 'fare':

```python
# Load the titanic dataset
titanic = sns.load_dataset("titanic")
titanic.head()
```

```python
# info of the DataFrame
titanic.info()
```

```python

# Plot a box plot for age vs. pclass
sns.boxplot(x="pclass", y="age", data=titanic)
plt.show()
```

```python

# Plot a box plot for fare vs. pclass
sns.boxplot(x="pclass", y="fare", data=titanic)
plt.show()
```

This code uses the `load_dataset()` function from Seaborn to load the "titanic" dataset. Then, two `boxplot()` functions are used to create box plots with x-axis as "pclass" and y-axis as "age" and "fare", respectively.

The resulting plots will show the distribution of the age and fare for each passenger class. The box represents the interquartile range (IQR), while the whiskers represent the range of data points within 1.5 times the IQR. Any points outside of the whiskers are considered outliers and plotted as individual points.

# Q6. 

Use the "flights" dataset from seaborn to plot a heatmap.


## Answer

Here's the code to load the "flights" dataset from Seaborn and plot a heatmap:

```python
# Load the flights dataset
flights = sns.load_dataset("flights")
flights.head()
```

```python
flights.info()
```

```python

# Pivot the data to create a matrix of flights by month
flights_pivot = flights.pivot(index="month", columns="year", values = "passengers")
flights_pivot
```

```python
flights_pivot.info()
```

```python

# Plot a heatmap of the flights data
sns.heatmap(flights_pivot, cmap="YlGnBu")
```

This code uses the `load_dataset()` function from Seaborn to load the "flights" dataset. Then, the `pivot()` function is used to create a matrix of flights by month. Finally, the `heatmap()` function is used to create a heatmap of the flights data.

The resulting plot will show the number of passengers on flights for each month and year. The color of each cell represents the number of passengers, with darker colors indicating a higher number of passengers. The `cmap` parameter sets the color palette of the heatmap.
